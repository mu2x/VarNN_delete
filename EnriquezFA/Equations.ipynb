## Equations

2.1. Conservative form of equations for reacting flows. Equations describing chemically reactive flows with N participating species in conservative formulation are stated as follows:

* Mass Conservation for Chemical Species:$$\dfrac{\partial\rho Y_i}{\partial t}+\nabla\cdot(\rho Y_i \mathbf{v})=-\nabla\cdot \mathbf{j}_i +\omega_i\quad (i=1,2,\cdots N)$$
* Mass Conservation for Mixture Gases:$$\dfrac{\partial \rho}{\partial t} +\nabla\cdot(\rho \mathbf{v})=0$$
* Conservation of Momentum: $$\dfrac{\partial \rho \mathbf{v}}{\partial t}+\nabla\cdot(\rho\mathbf{v}\otimes\mathbf{v})=-\nabla p + \nabla\cdot\boldsymbol{\tau}+\rho\sum_{i}^{N} Y_i \mathbf{f}_i$$
* Conservation of Energy:$$\rho\dfrac{\partial\rho e_t}{\partial t}+\nabla\cdot[(\rho e_t+p)\mathbf{v})]=-\nabla\cdot\mathbf{q}+\nabla\cdot(\boldsymbol{\tau}\cdot\mathbf{v})+\rho\sum_{i}^{N} Y_i\mathbf{f}_i\cdot\mathbf{v}+\sum_{i}^{N} \mathbf{f}_i\cdot\mathbf{j}_i$$ $$e_t=h-\frac{p}{\rho}+\frac{1}{2}\mathbf{v}\cdot\mathbf{v}$$
* Thermodynamic Equation of State:$$p=\rho R T M^{-1}\quad,\quad M=\biggl(\sum_{i=1}^{N} \frac{Y_i}{M_i}\biggr)^{-1}$$

where $\rho$ means the density, $\mathbf{v}$ denotes the velocity vector, $p$ stands for the pressure, $\boldsymbol{\tau}$ represents the viscous tensor, $\mathbf{f}_i$ means the body force per unit mass of species $i$, $Y_i$ represents the mass fraction of chemical species $i$, $\mathbf{j}_i$ denotes the diffusive flux vector of species $i$, $w_i$ stands for the mass production rate of species $i$, $e_t$ is the total energy, $q$ denotes the heat flux vector, $h$ represents the enthalpy, $R$ is the universal gas constant, $T$ denotes the temperature,
$M$ means the mean molecular mass, and $M_i$ stands for the molecular mass of species $i$. The
viscous stress tensor $\boldsymbol{\tau}$, the diffusive flux vector of species $\mathbf{j}_i$, and the heat flux vector $q$ will be given in the section of “Constitutive equation”.

### 1D EULER EQUATIONS
Assuming a fluid viscosity of zero in the Navier-Stokes equations yield the following. 
* Momentum Equation:$$\rho\biggl(\dfrac{\partial}{\partial t}u+u\dfrac{\partial}{\partial x}u\biggr)+\dfrac{\partial}{\partial x }p=g(x,t)$$
* Continuity Equation:$$\dfrac{\partial}{\partial t}\rho+\dfrac{\partial}{\partial x}(\rho u)=0$$
* Energy Equation:$$\dfrac{\partial}{\partial t}E+\dfrac{\partial}{\partial x}((E+p)u)=0$$
* Equations of State:$$p=c^2\rho\quad\text{(Acoustics - Sound waves)}$$ $$p=(\gamma-1)(E-\frac{1}{2}\rho u^2)\quad\text{(Gas Dynamics)}$$

Using the product rule

$$(\rho u)_t=\rho_t u+\rho u_t $$
Rearanging 

$$\rho u_t=(\rho u)_t-\rho_t u$$

Plugging $\rho_t$ from the continuity equation $\rho_t=-\dfrac{\partial}{\partial x}(\rho u)$ yields:

$$\rho u_t=(\rho u)_t+u \dfrac{\partial}{\partial x}(\rho u)$$

Plug the above $\rho u_t$ in the first term of the momentum equation

$$(\rho u)_t+u \dfrac{\partial}{\partial x}(\rho u)+\rho u\dfrac{\partial}{\partial x}u+\dfrac{\partial}{\partial x}p=$$
$$(\rho u)_t+\dfrac{\partial}{\partial x}(u\rho u)+\dfrac{\partial}{\partial x}p=g$$

This can be simplified to the following form of the 1D momentum equation:

$$(\rho u)_t+\dfrac{\partial}{\partial x}(u^2\rho+p)=g$$

For model propagation of sound, plug-in the Acoustics - Sound waves equation of state $p=c^2\rho$ in the momentim equation.

$$\dfrac{\partial}{\partial t}(\rho u)+\dfrac{\partial}{\partial x}((u^2+c^2)\rho)=g$$

substuting $\rho$ in the above yields the following form of the momentum equation:

$$ \dfrac{\partial}{\partial t}\biggl(\frac{p}{c^2}u\biggr)+\dfrac{\partial}{\partial x}\biggl(\frac{u^2+c^2}{c^2} p\biggr)=g$$

Doing the same $\rho$ substitution yields the following form of the continuity equation:

$$\dfrac{\partial}{\partial t}\frac{p}{c^2}+\dfrac{\partial}{\partial x}\biggl(\frac{p}{c^2}u \biggr)=0$$

### Linearizaion

Since particle velocity $u$ is small for sound waves, it is assumed that $u^2$ << $u$ and it can be dropped from the momentum and continuity equations. Also, assume no external body forces in the fluid.  

Momentum Equation

$$\dfrac{\partial}{\partial t}\biggl(\frac{p}{c^2}u\biggr)+\dfrac{\partial}{\partial x}p=0$$

Continuity Equation

$$\dfrac{\partial}{\partial t}\frac{p}{c^2}+\dfrac{\partial}{\partial x}\biggl(\frac{p}{c^2}u \biggr)=0$$
